```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def myPlot(x, y1, y1Name, y2, y2Name, xLabel, yLabel):
    plt.plot(x, y1, 'r', x, y2, 'b')
    plt.legend([y1Name,y2Name], loc = 'upper right')
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.show()
```

```python
layers = [4,6,12,20]

#Epoch = 50, training vs test
epoch50_train, epoch50_test = [0.8041, 0.8105, 0.8035, 0.7937], [0.763, 0.7602, 0.7857, 0.7599]
myPlot(layers, epoch50_train, "Epoch50 training", epoch50_test, "Epoch50 test", "Layers", "Accuracy")
```

```python
#Epoch = 100, training vs test
epoch100_train, epoch100_test = [0.8859, 0.8895, 0.8828, 0.8857], [0.8529, 0.8534, 0.8515, 0.8544]
myPlot(layers, epoch100_train, "Epoch100 training", epoch100_test, "Epoch100 test", "Layers", "Accuracy")
```

```python
#Epoch = 50 vs Epoch = 100 (test)
myPlot(layers, epoch50_test, "Epoch50 test", epoch100_test, "Epoch100 test", "Layers", "Accuracy")
```

```python
def scatterPlot(xList, yList, c, legend):
    plt.plot(xList[0], yList[0], color = c, label = legend)
    for x,y in zip(xList, yList):
        plt.plot(x, y, 'D', color = c)
```

```python
layers = [4,6,12,20]
scatterPlot(layers, epoch50_test, 'r', "Epoch = 50")
scatterPlot(layers, epoch100_test, 'b', "Epoch = 100")
layers, epoch200_test = [20, 32, 44, 56, 110, 164], [0.9125, 0.9249, 0.9283, 0.9303, 0.9339, 0.9407]
scatterPlot(layers, epoch200_test, 'g', "Epoch = 200")
plt.legend()
plt.xlabel("Layers")
plt.ylabel("Accuracy")
plt.show()
```
